```python
df = pd.read_csv(
    '/home/bergsmf/Projetos/Python/ECommerce/data/src/ecommerce-data.csv',
    delimiter=',',
    encoding='ISO-8859-1',
)
df
```

```python
df['InvoiceNo'] = pd.to_numeric(df['InvoiceNo'], errors='coerce').astype(
    'Int64'
)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').astype('Int64')
df['InvoiceDate'] = pd.to_datetime(
    df['InvoiceDate'], format='%d/%m/%Y %H:%M', errors='coerce'
)
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
df['CustomerID'] = pd.to_numeric(df['CustomerID'], errors='coerce').astype(
    'Int64'
)
df.dtypes
```
